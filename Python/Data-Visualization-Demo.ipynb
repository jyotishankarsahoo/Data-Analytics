## Data Visualization

### Histogram - (Distribution Of Individual Varibale)

**Histogram Of Price**

```python
sns.histplot(data= auto_mobile_data,
             x="price");
```

**Customize Histogram**

```python
plt.title("Histogram:Price")
plt.xlim(3000, 50000)
plt.ylim(0, 70)
plt.xlabel("Price Of Car")
plt.ylabel("Frequency")
sns.histplot(data= auto_mobile_data,
             x="price",
             color="Orange");
```

**Customize Bins**

```python
sns.histplot(data= auto_mobile_data,
             x="price",
             bins= 5,
             color="Orange");
```

```python
sns.histplot(data= auto_mobile_data,
             x="price",
             binwidth= 1000,
             color="Orange");
```

**Calculate Optimal Bin Width**

```python
# bin_width = 2 * IQR / Cuberoot(n)
# bins = Range / Bandwidth
```

**Kernel Density Estimation** (KDE)

```python
sns.histplot(data= auto_mobile_data,
             x="price",
             kde=True);
```

**Check Histogram For Few More Attributes**

```python
sns.histplot(data= auto_mobile_data,
             x="curb_weight",
             kde=True);
```

```python
sns.histplot(data= auto_mobile_data,
             x="horsepower",
             kde=True);
```

**Comparing distribution vs several groups**

```python
sns.histplot(data= auto_mobile_data,
             x = "price",
             hue = "body_style",
             kde= True);
```

### Line Plot (Sesonality Or Trend Of Data Over Time)

```python
flights_data_set = sns.load_dataset("flights")
sns.lineplot(data= flights_data_set,
             x= "month",
             y= "passengers");
# Shade is Confidence interval
# Solid line is avg value
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data= flights_data_set,
             x= "month",
             y= "passengers",
             errorbar=('ci', False),
             hue = "year")
plt.legend(bbox_to_anchor=[1,1])
plt.ylabel("No Of Passnager");
```

```python
fmri = sns.load_dataset("fmri")
sns.lineplot(data= fmri,
             x= "timepoint",
             y= "signal",
             errorbar=('ci', False),
             hue = "region",
             style="region");
```

```python
fmri = sns.load_dataset("fmri")
sns.lineplot(data= fmri,
             x= "timepoint",
             y= "signal",
             errorbar=('ci', False),
             hue = "region",
             style="region",
             markers= True);
```
