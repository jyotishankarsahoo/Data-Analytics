## NumPy Arrays

```python
# All Items Of NumpyArray are of same data type
att_str = ["Mercedes", "BMW", "Audi", "Ferrari", "Tesla"]
arr_num = [5, 4, 7, 3]
```

```python
# Converting list to NumpyArray
np_array_str = np.array(att_str)
np_array_num = np.array(arr_num)

print(f"Numpy Array (arr_str): {np_array_str} : of {type(np_array_str)}")
print(f"Numpy Array (arr_num): {np_array_num} : of {type(np_array_num)}")
```

```python
# Numpy Matrix
matrix = np.array([[1,2,1], [4,5,9], [1,8,9]])
print(f"Matix {matrix} Is of type: {type(matrix)}")
```

## Numpy Functions

**arange function**

```python
# arange Function
arr_step_2 = np.arange(start= 0,
                 stop= 10,
                 step= 2)
arr_step_5 = np.arange(start= 0,
                 stop= 20,
                 step= 5)
arr_step_default = np.arange(start= 0, # Default Step Size = 1
                 stop= 10)
print(f"arr_step_2: {arr_step_2}")
print(f"arr_step_5: {arr_step_5}")
print(f"arr_step_default: {arr_step_default}")
```

**linspace function**

```python
# linspace Function
matrix_20 = np.linspace(start= 0,
                       stop= 100,
                       num = 20) # Deafult value 50
print(f"20 Equally Distributed samples between 0 and 100: {matrix_20}")
print(f"Sample Step Size: {(100 - 0) / 19}") # (stop - start) / num_of_samples - 1
```

**zeros and ones function**

```python
# Zeros and Ones Function
matrix_zero = np.zeros(shape= [3,5]) # Default dType float
matrix_zero_int = np.zeros(shape= [3,5],
                           dtype= int)
matrix_one = np.ones(shape= [3,5])  # Default dType float
matrix_one_int = np.ones(shape= [3,5],
                           dtype= int)
print("3 rows and 5 coluns filled with Zeros:", matrix_zero)
print("3 rows and 5 coluns filled with Zeros:", matrix_zero_int)
print("3 rows and 5 coluns filled with Ones:", matrix_one)
print("3 rows and 5 coluns filled with Ones:", matrix_one_int)
```

**eye function**

```python
# Eye Function For Identity Matrix
# - matrix is squre with all zeros with diagonals ones
matrix_identity = np.eye(N= 5,
                         dtype=int)
print(f"Identity Matrix [5x5]: {matrix_identity}")
```

**reshape method**

```python
# Reshape Function
array_1 = np.arange(0,10)
reshaped_array_1 = array_1.reshape([2,5])
print(f"Reshaped Array: {reshaped_array_1}")
```

**Trignomtric Functions**

```python
print(f"Sine Function: {np.sin(4)}")
print(f"Cosine Function: {np.cos(4)}")
print(f"Tan Function: {np.tan(4)}")
```

**Exponent & Logarithmic Function**

```python
print(f"Exponent Num: {np.exp(2)}")
print(f"Exponent NumpyArray: {np.exp(np.arange(2,6,2))}")
```

```python
print(f"Log_e Num: {np.log(2)}")
print(f"Log_10 Num: {np.log10(100)}")
print(f"Log_2 Num: {np.log2(8)}")
print(f"Log_e NumpyArray: {np.log(np.arange(2,6,2))}")
print(f"Log_10 NumpyArray: {np.log10(np.arange(2,6,2))}")
```

**Arithmetic Operation**

```python
p_array1 = [1,2,3]
p_array2 = [4,5,6]
print(f"Python Array Sum: {p_array1 + p_array2}")

np_array_1 = np.array(p_array1)
np_array_2 = np.array(p_array2)
print(f"NumPy Array Sum: {np_array_1 + np_array_2}")
print(f"NumPy Array Sub: {np_array_1 - np_array_2}")
print(f"NumPy Array Mul: {np_array_1 * np_array_2}")
print(f"NumPy Array Dev: {np_array_1 / np_array_2}")
print(f"NumPy Array Powers: {np_array_1 ** np_array_2}")
```

```python
# Operation on Matrix
matrix_1 = np.arange(start=1, stop=10).reshape(3,3)
matrix_2 = np.eye(3, dtype=int)
matrix_3 = np.ones(shape= (3,3), dtype=int)

print(matrix_1)
print(matrix_2)
print(matrix_3)
```

```python
print(f"Matrix Addtion: \n {matrix_1 + matrix_2}")
print(f"Matrix Sub: \n {matrix_1 - matrix_2}")
print(f"Matrix mul: \n {matrix_1 * matrix_2}")
print(f"Matrix div: \n {matrix_1 / matrix_3}")
print(f"Matrix Linear mul: \n {matrix_1 @ matrix_3}")
```

```python
# Transpose of matrix
matrix_for_trans = np.arange(1, 11).reshape([2,5])
matrix_trans = np.transpose(matrix_for_trans)
print(f"Original Matrix: \n{matrix_for_trans}")
print(f"Transpose Matrix: \n{matrix_for_trans.T}")
print(f"Transpose Matrix: \n{matrix_trans}")
```

**Min Max Functions**

```python
vec = np.arange(1,10).reshape([3,3])
print("Max value:", np.max(vec))
print("Min value:", np.min(vec))
```

**Random Function**

```python
rand_mat = np.random.rand(5)
rand_mat_uniform = np.random.rand(5,5)
rand_mat_n = np.random.randn(5,5)
rand_int = np.random.randint(0, 5, 10)
rand_mat_int = np.random.randint(0, 10, [5,5])
print("Random Matrix [1x5]: ", rand_mat)
print("Random Matrix Int [5x5]: ", rand_mat_n)
print("Random Matrix [5x5]: ", rand_mat_uniform)
print("Random Int [1x5]: ", rand_int)
print("Random Int [2x2]: ", rand_mat_int)
```

**Mean & Median**

```python
print("Mean: ",np.mean(rand_mat_n) )
print("Std: ",np.std(rand_mat_n) )
```

## Accessing Numpy Array

```python
# Accessing Array
rand_arry = np.random.rand(10)
print(f"rand_array: {rand_arry}")
print(f"rand_array[0:4]: {rand_arry[0:4]}")
print(f"rand_array[rand_arry[np.arange(3,10,3]: {rand_arry[np.arange(3,10,3)]}")
print(f"rand_array[rand_arry>0]: {rand_arry[rand_arry > 0]}")
print(f"rand_array[rand_arry>0]: {rand_arry[rand_arry < 0]}")
```

```python
# Accessing Matrix
rand_m = np.random.randn(5, 5)
print("Matrix 5x5: \n", rand_m)
print("Matrix[1,4] - 1st row and 4ith column: \n", rand_m[1,4])
print("Matrix[2] - 2nd row: \n", rand_m[2])
print("Matrix[0:2, 1:3] - 0 and 2nd row with 1 and 3 column: \n", rand_m[0:2, 1:3])
```

## Modifying NumPy Array

```python
print(rand_arry)
# Change values of index 3 and 4 to 5
rand_arry[3:5] = 5
print("Values of index 3 and 4 to 5",rand_arry)
# Change value of index 0 and 1 to 2 and 3
rand_arry[0:2] = [2,3]
print("Values of index 0 and 1 to 2,3: ",rand_arry[0],rand_arry[1])
```

## Modifying NumPy Matrix

```python
rand_mat_3 = np.random.randint(1, 10, [5,5])
print("Before 5x5 Matrix: \n", rand_mat_3)
rand_mat_3[1:3, 3:5] = 0
print("After 5x5 Matrix: \n", rand_mat_3)
# Extracting first 2 rows and first 3 column
sub_matrix = rand_mat_3[0:2, 0:3].copy()
print("Sub-Matrix: \n", sub_matrix)
sub_matrix[:] = 3
print("Sub-Matrix Updated: \n", sub_matrix)
```
